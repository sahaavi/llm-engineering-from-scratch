## Build It

Byte-level BPE begins with one token for each possible byte. Training repeatedly counts adjacent pairs, merges the most frequent pair, and adds that merged byte sequence to the vocabulary.

```python
from main import BPETokenizer, TINY_CORPUS, STRESS_EXAMPLES, build_artifacts

tokenizer = BPETokenizer().train(TINY_CORPUS, vocab_size=286)
text = "tokenization turns text into token ids."
token_ids = tokenizer.encode(text)

print(token_ids)
print([piece["text"] for piece in tokenizer.token_pieces(token_ids)])
print(tokenizer.decode(token_ids))
```
